## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading The Dataset

```python
data = pd.read_csv("apple.csv")
data.head(5)
```

```python
data.info()
df = data.iloc[:, [0, 1]]
df
df.set_index('Year', inplace=True)
```

```python
sns.lineplot(df)
```

```python
sns.heatmap(df)
```

```python
df['Value'].describe()
```

```python
df.isnull().sum()
```

```python
plt.plot(df.index, df["Value"])
plt.ylabel('Price')
plt.xlabel('Date')
plt.title('Opening Price of the Stocks')
plt.xticks(rotation =45)
plt.xlim(df.index.min(), df.index.max())
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
DR = seasonal_decompose(df["Value"], model = 'multiplicative', period=3)
DR.plot()
plt.show()
```
