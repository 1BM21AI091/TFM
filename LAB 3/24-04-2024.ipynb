```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
np.random.seed(0)
white_noise = np.random.normal(size=309)

sunspots_data = sm.datasets.sunspots.load_pandas().data['SUNACTIVITY']
```

```python
plt.figure(figsize=(10, 5))
plt.plot(white_noise, color='blue', label='White Noise')
plt.plot(sunspots_data, color='green', label='Sunspots Data')
plt.title('Comparison of White Noise and Sunspots Data')
plt.xlabel('Year')
plt.ylabel('Sunspot Activity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
adf_result = adfuller(sunspots_data)
print('Augmented Dickey-Fuller Test:')
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])
if adf_result[1] < 0.05:
    print('The sunspots data is likely stationary (reject the null hypothesis)')
else:
    print('The sunspots data is likely non-stationary (fail to reject the null hypothesis)')
```

```python
kpss_result = kpss(sunspots_data)
print('\nKwiatkowskiPhillipsSchmidtShin Test:')
print('KPSS Statistic:', kpss_result[0])
print('p-value:', kpss_result[1])
print('Critical Values:', kpss_result[3])
if kpss_result[1] > 0.05:
    print('The sunspots data is likely stationary (fail to reject the null hypothesis)')
else:
    print('The sunspots data is likely non-stationary (reject the null hypothesis)')
```
